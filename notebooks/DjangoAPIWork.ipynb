```python
import requests
import json
baseURL="http://b.libtech.in"
apiusername='apimanager'
apipassword='sharedata'
AUTHENDPOINT='%s/auth/jwt/' % (baseURL)
TESTURL="%s/api/test/" % (baseURL)
REPORTURL="%s/api/report/" % (baseURL)
GETREPORTURL="%s/api/getReport/" % (baseURL)
LOCATIONURL="%s/api/location/" % (baseURL)
CHATTERPUR_BARA="3405008014"
CHATTERPUR_BLOCK="3405008"
JAVAJA_BLOCK="2721003"
JOBCARD='RJ-272100309802522500/263'
itdaBlocks=['0203006','0203005','0203012','0203004','0203011','0203013','0203003','0203014','0203001','0203010','0203002']
```

```python
def getHeaders():
  data={
      'username' : apiusername,
      'password' : apipassword
            }
  r=requests.post(AUTHENDPOINT,data=data)
  token=r.json()['token']
  headers={
      'content-type':'application/json',
      "Authorization" : "JWT " + token
    }
  return headers
```

```python
def getAllDistrict():
  locationArray=[]
  locationType='district'
  scheme='nrega'
  url=f"{LOCATIONURL}?locationType={locationType}&scheme={scheme}&limit=10000"
  r=requests.get(url)
  response=r.json()
  results=response['results']
  for res in results:
    dCode=res['code']
    locationArray.append(dCode)
  return locationArray
```

```python
def initiateCrawl(locationCode=None,priority=100,scheme='nrega',reportType=None,iterateChild=False,startFinYear='19'):
  if (locationCode is None) or (reportType is None):
    return
  lArray=[]
  if iterateChild:
    url=f"{LOCATIONURL}?parentLocation__code={locationCode}&scheme={scheme}&limit=10000"
    r=requests.get(url)
    response=r.json()
    results=response['results']
    for res in results:
      dCode=res['code']
      lArray.append(dCode)
  else:
    lArray.append(locationCode)
  for lCode in lArray:
    url=f"{GETREPORTURL}?reportType={reportType}&location__code={lCode}&scheme={scheme}&priority={priority}&startFinYear={startFinYear}"
    print(url)
    headers=getHeaders()
    r=requests.get(url,headers=headers)
    print(r.json())
```

```python
reportType="jobcardRegister"
priority=150
scheme='nrega'
itdaBlocks=['0203006','0203005','0203012','0203004','0203011','0203013','0203003','0203014','0203001','0203010','0203002']
for locationCode in itdaBlocks:
  initiateCrawl(locationCode=locationCode,priority=priority,scheme='nrega',reportType=reportType,iterateChild=True,startFinYear='19')
```
